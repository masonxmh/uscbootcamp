```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# drop repeat SN
sns = purchase_data["SN"].unique()
#calculate total number of players
total_players = len(sns)
#create data frame for display
total_players_df=pd.DataFrame({"Total Players":[total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# number of unique items
uniq_items = purchase_data["Item ID"].unique()
number_uniq_items = len(uniq_items)

# Total revenue
sum_price = purchase_data["Price"].sum()
sum_price = '${:,.2f}'.format(sum_price)

#Average Price
ave_price = purchase_data["Price"].mean()
ave_price ='${:,.2f}'.format(ave_price)
#number of Purchases
num_purchases = purchase_data["Purchase ID"].count()


#display data frame
Purchasing_Analysis=pd.DataFrame({"Number of Unique Items":[number_uniq_items],
                                  "Average Price":[ave_price],
                                  "Number of Purchases":[num_purchases],
                                  "Total Revenue":[sum_price]})
Purchasing_Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#drop repeat SN
gd = purchase_data.drop_duplicates("SN")
#calculate total players
gd = gd.groupby(['Gender'])
total_player=gd["SN"].count()

# Calculate percentage of players
gd_percent = total_player/total_player.sum()*100

# Generate the DataFrame for display
gd_df = pd.DataFrame({"Total Count": total_player,
                      "Percentage of Players": gd_percent})
gd_df["Percentage of Players"] = gd_df["Percentage of Players"].map("{:.2f}%".format)
gd_df = gd_df.sort_values("Total Count", ascending=False)
gd_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Check maximum value
print(purchase_data["Age"].max())

#Generate Bins 
bins = [0,9,14,19,24,29,34,39,50]
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
# Set up bins
purchase_data["Age Group"]=pd.cut(purchase_data["Age"],bins,labels=group_labels)
#Calculate percentage of players
age= purchase_data.groupby(['Age Group'])["SN"].nunique()
age_total = age.sum()
age_percent = age/age_total*100
#Generate display dataframe
age_table=pd.DataFrame({"Total Count":age, "Percentage of Players":age_percent}) 
age_table.index.name=None
age_table["Percentage of Players"]=age_table["Percentage of Players"].map("{:.2f}%".format)
age_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#set up bins
bins = [0,9,14,19,24,29,34,39,50]
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data["Age Group"]=pd.cut(purchase_data["Age"],bins,labels=group_labels)
age_purchase= purchase_data.groupby(['Age Group'])

age_purchase_count = age_purchase["Age"].count()


# average purchase price 
age_ave_purchase_price = age_purchase["Price"].mean()


# total purchase value 
age_total_purchase_value=age_purchase["Price"].sum()

#ave total purchase per person
age_total_perperson= age_total_purchase_value/age
# Generate DataFrame
age_purchase_table = pd.DataFrame({"Purchase Count":age_purchase_count, 
                                   "Average Purchase Price":age_ave_purchase_price,
                                   "Total Purchase Value":age_total_purchase_value,
                                   "Avg Total Purchase per Person":age_total_perperson})
#Format DataFrame
age_purchase_table.index.name=None
age_purchase_table["Average Purchase Price"]=age_purchase_table["Average Purchase Price"].map("${:.2f}".format)
age_purchase_table["Total Purchase Value"]=age_purchase_table["Total Purchase Value"].map("${:,.2f}".format)
age_purchase_table["Avg Total Purchase per Person"]=age_purchase_table["Avg Total Purchase per Person"].map("${:.2f}".format)
age_purchase_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#calculate results
spender=purchase_data.groupby(['SN'])
spender_purchase_count = spender["Purchase ID"].count()
spender_purchase_ave = spender["Price"].mean()
spender_purchase_total = spender["Price"].sum()
#Generate dataframe
spender_purchase_table = pd.DataFrame({"Purchase Count":spender_purchase_count,
                                       "Average Purchase Price":spender_purchase_ave,
                                       "Total Purchase Value":spender_purchase_total})

#format dataframe
sort_spender_df=spender_purchase_table.sort_values("Total Purchase Value", ascending=False)

sort_spender_df["Average Purchase Price"]=sort_spender_df["Average Purchase Price"].map("${:.2f}".format)
sort_spender_df["Total Purchase Value"]=sort_spender_df["Total Purchase Value"].map("${:.2f}".format)


sort_spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
popular_items=purchase_data.groupby(['Item ID','Item Name'])
item_price =popular_items["Price"].unique()
pitem_count = popular_items['Item Name'].count()
price_sum=popular_items["Price"].sum()

#generate display dataFrame
popular_table=pd.DataFrame({"Purchase Count":pitem_count,
                            "Item Price":item_price,
                            "Total Purchase Value":price_sum})
#format dataframe
sort_popular_table=popular_table.sort_values("Purchase Count", ascending=False)
sort_popular_table["Item Price"]=sort_popular_table["Item Price"].astype(float)
sort_popular_table["Item Price"]=sort_popular_table["Item Price"].map("${:.2f}".format)
sort_popular_table["Total Purchase Value"]=sort_popular_table["Total Purchase Value"].map("${:.2f}".format)
sort_popular_table.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Sort the above table by total purchase value in descending order
sort_popular_table=popular_table.sort_values("Total Purchase Value", ascending=False)
#Format display dataframe
sort_popular_table["Item Price"]=sort_popular_table["Item Price"].astype(float)
sort_popular_table["Item Price"]=sort_popular_table["Item Price"].map("${:.2f}".format)
sort_popular_table["Total Purchase Value"]=sort_popular_table["Total Purchase Value"].map("${:.2f}".format)
sort_popular_table.head()
```
