```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
school_column = school_data_complete["School ID"].unique()
total_schools = len(school_column)

#Calculate the total number of students
total_students = school_data_complete["Student ID"].count()

#Calculate the total budget
total_budget = school_data_complete["budget"].unique().sum()

#Calculate the average math score
ave_mathscore= school_data_complete["math_score"].mean()

#Calculate the average reading score
ave_readingscore = school_data_complete["reading_score"].mean()

#Calculate the percentage of students with a passing math score (70 or greater)
percent_math = len(school_data_complete.loc[school_data_complete["math_score"] >=70])/total_students*100

#Calculate the percentage of students with a passing reading score (70 or greater)
percent_reading = len(school_data_complete.loc[school_data_complete["reading_score"] >=70])/total_students*100

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
percent_overall = (ave_mathscore+ave_readingscore)/2

#Generate the DataFrame for summary
Distrct_Table = pd.DataFrame({"Total Schools":[total_schools],
                              "Total Students":[total_students],
                             "Total Budget":[total_budget],
                             "Average Math Score":[ave_mathscore],
                             "Average Reading Score":[ave_readingscore],
                              "% Passing Math":[percent_math],
                             "% Passing Reading":[percent_reading],
                              "% Overall Passing Rate":[percent_overall]})
#Change the data format

Distrct_Table["Total Students"]=Distrct_Table["Total Students"].map("{:,}".format)
Distrct_Table["Total Budget"]=Distrct_Table["Total Budget"].map("${:,.2f}".format)
# Distrct_Table["Average Math Score"]=Distrct_Table["Average Math Score"].map("{:.2f}".format)
# Distrct_Table["Average Reading Score"]=Distrct_Table["Average Reading Score"].map("{:.2f}".format)
# Distrct_Table["% Passing Math"]=Distrct_Table["% Passing Math"].map("{:.2f}".format)
# Distrct_Table["% Passing Reading"]=Distrct_Table["% Passing Reading"].map("{:.2f}".format)
# #Distrct_Table["% Overall Passing Rate"]=Distrct_Table["% Overall Passing Rate"].map("{:.2f}".format)
Distrct_Table
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_df = school_data_complete.groupby(['school_name'])

school_type=school_df["type"].unique().str[0]

total_students = school_df["Student ID"].count().astype(object)

total_school_budget = school_df["budget"].unique().astype(float)

per_student_budget = total_school_budget/total_students

ave_matchscore = school_df["math_score"].mean()
                                                        
ave_readingscore = school_df["reading_score"].mean()

percent_passmath =school_data_complete.loc[school_data_complete["math_score"] >=70]
percent_passmath = percent_passmath.groupby(['school_name'])["math_score"].count()/total_students


percent_passreading =school_data_complete.loc[school_data_complete["reading_score"] >=70]
percent_passreading = percent_passreading.groupby(['school_name'])["reading_score"].count()/total_students

percent_overall = (percent_passmath + percent_passreading)/2

school_table = pd.DataFrame({"School Type":school_type,
                             "Total Students":total_students,
                             "Total School Budget":total_school_budget,
                             "Per Student Budget":per_student_budget,
                             "Average Math Score":ave_matchscore,
                              "Average Reading Score": ave_readingscore,
                              "% Passing Math":percent_passmath*100,
                              "% Passing Reading":percent_passreading*100,
                              "% Overall Passing Rate":percent_overall*100})

school_table["% Passing Math"] = school_table["% Passing Math"].astype(float)
school_table["% Passing Reading"] = school_table["% Passing Reading"].astype(float)
school_table["% Overall Passing Rate"] = school_table["% Overall Passing Rate"].astype(float)

school_table_clean=school_table.copy()
school_table_clean["Total School Budget"] = school_table_clean["Total School Budget"].map("${:,.2f}".format)
school_table_clean["Per Student Budget"] = school_table_clean["Per Student Budget"].map("${:,.2f}".format)

school_table_clean.index.name=None
school_table_clean.head()

#percent_passmath/total_students
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
top_school_table=school_table_clean.sort_values("% Overall Passing Rate", ascending=False)
top_school_table.head(5)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

school_table["Per Student Budget"] = school_table["Per Student Budget"].astype(float)
school_table["Spending Ranges (Per Student)"]=pd.cut(school_table["Per Student Budget"],bins=spending_bins,labels=group_names)


budget_per_st_table=school_table.groupby(['Spending Ranges (Per Student)'])


budget_ave_math=budget_per_st_table["Average Math Score"].mean()
budget_ave_reading=budget_per_st_table["Average Reading Score"].mean()


budget_percent_math = budget_per_st_table["% Passing Math"].mean()
budget_percent_reading = budget_per_st_table["% Passing Reading"].mean()
budget_percent_overall = budget_per_st_table["% Overall Passing Rate"].mean()

budget_per_st_table = pd.DataFrame({"Average Math Score":budget_ave_math,
                                    "Average Reading Score":budget_ave_reading,
                                    "% Passing Math":budget_percent_math,
                                    "% Passing Reading":budget_percent_reading,
                                    "% Overall Passing Rate":budget_percent_overall})
budget_per_st_table
```

## Scores by School Size
